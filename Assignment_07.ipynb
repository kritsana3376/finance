```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
def plot2(X:np.array,
          Y:np.array,
          Y_predict:np.array,
          title):
  plt.figure(figsize = (15,10))
  plt.plot(X,Y_predict,label = "Regression Model")
  plt.plot(X,Y,"o",label = "Actual Value")
  plt.title(title)
  plt.xlabel("Year")
  plt.ylabel("EPS")
  plt.legend()
  plt.grid()
```

```python
# -> DELTA
lst = [2556, 2557, 2558, 2559, 2560, 2561, 2562, 2563, 2564, 2565]
# list of int 
lst1 = [ 4.34,  4.78,  5.38,  4.42,  3.95,  4.12,  2.37,  5.69,  5.37,12.3 ]

DELTA_Stocks = pd.DataFrame(list(zip(lst, lst1)), columns =['Year', 'EPS'])
DELTA_Stocks.to_csv("DELTA_STOCKS.csv")
```

```python
# -> COM7
lst2 = [ np.NaN, 0.23, 0.26, 0.34, 0.51, 0.74, 1.01, 1.24, 2.19,  np.NaN]
 
COM7_Stocks = pd.DataFrame(list(zip(lst, lst2)), columns =['Year', 'EPS'])
COM7_Stocks.to_csv("COM7_STOCKS.csv")
COM7_Stocks = COM7_Stocks.dropna()
```

```python
# -> TU
lst3 = [2.49, 1.1 , 1.11, 1.1 , 1.26, 0.68, 0.8 , 5.69, 1.26, 1.47]

TU_Stocks = pd.DataFrame(list(zip(lst, lst3)), columns =['Year', 'EPS'])
TU_Stocks.to_csv("TU_STOCKS.csv")
```

```python
#-> COM7 
COM7s_x = COM7_Stocks["Year"].to_numpy().reshape(-1,1) 
COM7s_y = COM7_Stocks["EPS"].to_numpy()
#-> DELTA
DELTAs_x = DELTA_Stocks["Year"].to_numpy().reshape(-1,1)
DELTAs_y = DELTA_Stocks["EPS"].to_numpy()
#-> TU
TUs_x = TU_Stocks["Year"].to_numpy().reshape(-1,1)
TUs_y = TU_Stocks["EPS"].to_numpy()
```

```python
#-> COM7 
reg_COM7 = LinearRegression().fit(COM7s_x,COM7s_y)
#-> DELTA
reg_DELTA = LinearRegression().fit(DELTAs_x,DELTAs_y)
#-> TU
reg_TU = LinearRegression().fit(TUs_x,TUs_y)
```

```python
y_pred_COM7 = reg_COM7.predict(COM7s_x)

y_pred_DELTA = reg_DELTA.predict(DELTAs_x)

y_pred_TU = reg_TU.predict(TUs_x)
```

```python
plot2(COM7s_x,COM7s_y,y_pred_COM7,"COM7")
```

```python
plot2(DELTAs_x,DELTAs_y,y_pred_DELTA,"DELTA")
```

```python
plot2(TUs_x,TUs_y,y_pred_TU,"TU")
```
